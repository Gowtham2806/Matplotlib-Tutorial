## Gas_Price Plot

```python
plt.figure(figsize=(8,6))

plt.plot(gas.Year, gas.USA, 'r*--', label='USA', linewidth=2, markersize=10, markeredgecolor='blue')
plt.plot(gas.Year, gas.Canada, label='Canada', linewidth=2, marker = '^', markersize=10)
plt.plot(gas.Year, gas['South Korea'], label='South Korea', linewidth=2, marker='o', markersize=10)

plt.title('Gas Prices Over Time (in USD)')
plt.xlabel("Year")
plt.ylabel("Price (in USD)")

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.savefig('Gas_Price.png', dpi=300)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#aacdef', '#acdbef']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')
plt.title("Foot Preferences of FIFA Players", fontdict={'fontname': 'Comic Sans Ms', 'fontsize': 15})

plt.show()
```

```python
# def Weight():
#     for x in fifa.Weight:
#         if type(x) == str:
#             return int(x.strip('lbs'))
#         else:
#             return x    

fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy= fifa.loc[(fifa.Weight >= 200)].count()[0]
```

```python
plt.figure(figsize=(8,6))

plt.style.use('ggplot')

labels = ['light', 'light_medium', 'medium', 'medium_heavy', 'heavy']

explode = (0.4, 0,0,0,0.4)

plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.pie([light, light_medium, medium, medium_heavy, heavy], labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.show()
```

## Box & Whiskers

```python
plt.style.use('default')
plt.figure(figsize=(6,6))

Manchester = fifa[fifa.Club == 'Manchester United']['Overall']
madrid = fifa[fifa.Club == 'Real Madrid']['Overall']
New_York = fifa[fifa.Club == 'New York City FC']['Overall']

labels = ['Manchester United', 'Real Madrid', 'New York City FC']
plots = [Manchester, madrid, New_York]
plt.boxplot(plots, labels = labels)

plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.xlabel('FIFA Clubs')

plt.show()
```
