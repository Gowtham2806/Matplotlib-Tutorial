```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
x = np.arange(1,4)
y = np.arange(2,7,2)

plt.figure(figsize=(6,4))

plt.plot(x, y, label='2x', color='red', linewidth = 2, marker = '.', markersize = 10, markeredgecolor='blue', linestyle='--')

plt.title("Our first graph", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X - Points', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.ylabel('Y - Points', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})

# plt.xticks([1,2,3])
# plt.yticks([2,4,6])

x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:6], x2[:6]**2, 'y', label='x2^2', linewidth=2.5, marker='^', markersize=10)
plt.plot(x2[5:], x2[5:]**2, 'y*--', label='x2^2', linewidth=2.5, markersize=10)

plt.legend()

plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = np.arange(2,7,2)

plt.figure(figsize=(6,4))

bars = plt.bar(labels, values)

patters = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patters.pop(0))
    
# bar[0].set_hatch('/')
# bar[1].set_hatch('*')
# bar[2].set_hatch('o')

plt.show()
```
